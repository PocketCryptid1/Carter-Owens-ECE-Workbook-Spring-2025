### Question: what is the difference between the performance of the straight_line path and the fillet path?

**Fillet Path:**
- The UAV follows smooth, curved transitions (circular arcs) between straight-line segments at each waypoint
- The path is continuous and differentiable, meaning the UAV doesn't need to make sharp turns
- This results in smoother flight dynamics with less aggressive roll commands
- The UAV doesn't pass directly through the waypoints but rather follows tangent arcs near them

**Straight-Line Path:**
- The UAV flies directly toward each waypoint in a straight line
- At waypoints, the UAV must make sharp turns to redirect toward the next waypoint
- This requires more aggressive maneuvering and higher roll angles at waypoint transitions
- The UAV passes through (or very close to) each waypoint before transitioning

**Performance Comparison:**
- The fillet path produces smoother tracking with less overshoot and oscillation in the course angle
- The straight-line path may exhibit more aggressive transients at waypoint transitions
- Fillet paths are generally preferred for fixed-wing aircraft because they respect the aircraft's turn radius constraints and produce flyable trajectories
- Straight-line paths are simpler to compute but may command infeasible maneuvers if waypoint angles are too sharp

````python
# Problem 3: Create 5 non-colinear waypoints starting and ending at (0, 0, -100)
# With sufficient spacing to prevent fillet circle overlap

# Create waypoints for fillet path first
waypoints = MsgWaypoints()
waypoints.type = 'fillet'
Va = PLAN.Va0

# Five waypoints: start at origin, make a path, return to origin
# Spacing of 500-1000m between waypoints to ensure fillets don't overlap
waypoints.add(np.array([[0, 0, -100]]).T, Va, np.radians(0), np.inf, 0, 0)      # Start
waypoints.add(np.array([[800, 400, -100]]).T, Va, np.radians(45), np.inf, 0, 0)  # NE
waypoints.add(np.array([[1200, -200, -100]]).T, Va, np.radians(-30), np.inf, 0, 0) # SE
waypoints.add(np.array([[600, -600, -100]]).T, Va, np.radians(-135), np.inf, 0, 0) # SW
waypoints.add(np.array([[0, 0, -100]]).T, Va, np.radians(180), np.inf, 0, 0)     # Return to start

# Run fillet path simulation
print("Running FILLET path simulation...")
run_sim_and_display(waypoints=waypoints)

# Now run straight-line path simulation with same waypoints
waypoints.type = 'straight_line'
waypoints.flag_waypoints_changed = True
print("\nRunning STRAIGHT-LINE path simulation...")
run_sim_and_display(waypoints=waypoints)
````
